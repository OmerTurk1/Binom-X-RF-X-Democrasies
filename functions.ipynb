```python
import math
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def factorial(n):
    return math.factorial(n)

def binom(n,p,r):
    return (factorial(n)/(factorial(r)*factorial(n-r))) * p**r * (1-p)**(n-r)
```

```python
def calculate(n,p):
    result = 0
    for r in range(int(np.ceil(n/2)),n+1):
        ans = binom(n,p,r)
        result += ans
    return result
```

```python
P = np.arange(0.50, 0.81, 0.01)
N = np.arange(3, 25, 1)
A = []
for p in P:
    for n in N:
        A.append([p,n,calculate(n,p)])
```

```python
x = [round(row[0],2) for row in A]
y = [row[1] for row in A]
z = [row[2] for row in A]

# Create matrix
x_unique = sorted(set(x))
y_unique = sorted(set(y))
z_matrix = np.zeros((len(y_unique), len(x_unique)))

for xi, yi, zi in zip(x, y, z):
    x_idx = x_unique.index(xi)
    y_idx = y_unique.index(yi)
    z_matrix[y_idx, x_idx] = zi

# Heatmap drawing
plt.figure(figsize=(8, 6))
sns.heatmap(z_matrix, xticklabels=x_unique, yticklabels=y_unique, cmap="Blues", vmin=0, vmax=1)
plt.xlabel("P accuracy")
plt.ylabel("N amount")
plt.title("Heatmap of Accuracy")
```
